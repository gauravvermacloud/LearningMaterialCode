```python
import numpy as np

import matplotlib.pyplot as plt
```

## Uniform Distribution

```python
uniform_distribution = np.random.uniform(-10,10,100000)
plt.hist(uniform_distribution,50)
plt.show()
```

```python
from scipy.stats import uniform
x = np.arange(-3,3,0.001)
plt.plot(x, uniform.pdf(x))
```

## Normal Distribution

```python
avg = 0
std_div = 1
values = np.random.normal(avg,std_div,100000)
plt.hist(values,50)
plt.show()
```

```python
from scipy.stats import norm
x = np.arange(-3,3,0.001)
plt.plot(x,norm.pdf(x))
```

## Exponential

```python

values = np.random.exponential(1, 100000)
plt.hist(values, 50)
plt.show()
```
