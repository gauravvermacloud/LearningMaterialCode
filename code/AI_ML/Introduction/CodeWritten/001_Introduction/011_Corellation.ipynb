```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(3,1,1000)
y = np.random.normal(50,10,1000)

plt.scatter(x,y)
```

```python
correlation = np.corrcoef(x,y)
correlation
#correlation between x,x x,y  y,x and x,x
```

```python
y = np.random.normal(50,10,1000) / x
plt.scatter(x, y)
plt.show()
```

```python
correlation = np.corrcoef(x, y)
correlation
```

```python
y = x*3
plt.scatter(x, y)
plt.show()
```

```python
correlation = np.corrcoef(x, y)
correlation
```
