```python
#Generate some data
import  matplotlib.pyplot as plt
import numpy as np

#Generate 1000 sanmples with mean of 3 and SD of 1
x = np.random.normal(3.0,1.0,1000)

#for each x value subtract a random valiu around 0 with SD = 0.1 and multiply this number with 3 and subtract it from 100
y = 100-(x + np.random.normal(0.0, 0.1, 1000))*3

#This will be a lineear relation
#Now lets plot it

plt.scatter(x,y)
plt.show()
```

```python
from scipy import stats

result = stats.linregress(x,y)
result
slope,intercept,rvalue,pvalue,stderr= result
print(result)

#Basicall m =  slope and c = intercept and we can make y = mx+c
#The err r squared = rvale**2
```

```python
#r squared
r_sqr = rvalue**2
r_sqr
```

```python
#now let us make the linear prediction function

def predict_linear(x):
    return x*slope +intercept
```

```python
#Predicting for some random value

predict_linear(2)
```

```python
#Now lets make a best fit line on our scatter plot
# We have x values let us get the y values for it

y_predicted = predict_linear(x)
print(x)
print(y_predicted)
```

```python
#Now lets make a scatter and line plot
plt.scatter(x,y)
plt.plot(x, y_predicted,color='red')
plt.show()
```

```python
#doing the same using a 1 d polynomial 

poly_fit_1st_degree_function = np.poly1d(np.polyfit(x, y, 1))
```

```python
y1 = predict_linear(5)
y2 = poly_fit_1st_degree_function(5)

print(y1)
print(y2)
```
