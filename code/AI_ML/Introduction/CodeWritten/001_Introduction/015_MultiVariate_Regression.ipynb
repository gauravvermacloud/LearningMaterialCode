```python
#Now we want to predict price as a function of Milage, Cylender and Door
#Regression cant be done on catagorical data

import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()


X = df[['Mileage', 'Cylinder', 'Doors']]
y = df['Price']
X = df[['Mileage', 'Cylinder', 'Doors']] = scale.fit_transform(X[['Mileage', 'Cylinder', 'Doors']])
print(X)
print(y)
```

```python
#Let us normalize the input values of x to have values between -1 and 1
est = sm.OLS(y, X).fit()

est.summary()
```
