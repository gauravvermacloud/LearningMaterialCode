```python
x = np.arange(-3,3,0.01)
x
```

```python
from  scipy.stats import norm
y1 = norm.pdf(x)
y2 = norm.pdf(x,1.0,0.5)
y1
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```
