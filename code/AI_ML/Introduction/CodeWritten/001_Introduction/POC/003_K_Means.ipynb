```python
import numpy as np
import pandas as pd
#Lets load the data
col_name = ['sepal-lenght', 'sepal-width',
            'petal-lenght', 'petal-width', 'class']

df = pd.read_csv('./iris.csv', names=col_name)

df.head

```

```python
X = df[['sepal-lenght', 'sepal-width',
        'petal-lenght', 'petal-width']]
X
```

```python
#Lets do all  plots

import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(X, markers='+')
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(X)
#Visually we can see 2 classes coming out
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
print(kmeans.labels_)
```

```python
kmeans.predict([[6.2,3.4,5.4,2.3]])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=['a','b'])
plt.show()
```
