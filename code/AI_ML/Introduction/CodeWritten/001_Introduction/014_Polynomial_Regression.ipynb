```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

x = np.random.normal(3,1.0,1000)
y = np.random.normal(100,10,1000)/x

plt.scatter(x,y)
plt.show()
```

```python
# Let us assume this is fit well using a 4th degree polynomial, so we create the function

poly_fit_4th_degree_function = np.poly1d(np.polyfit(x,y,4))
```

```python
print(poly_fit_4th_degree_function(4))
print(poly_fit_4th_degree_function(1))
```

```python
#finally lets build the curve


plt.scatter(x, y)

x_predicted = np.linspace(0,7,700)
y_predicted = poly_fit_4th_degree_function(x_predicted)
plt.plot(x_predicted, y_predicted, color='red')
plt.show()
```

```python
#Finally getting r squared
from sklearn.metrics import r2_score

r_squared = r2_score(y, poly_fit_4th_degree_function(x))

r_squared
```
